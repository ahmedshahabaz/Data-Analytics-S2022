```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier 
from sklearn.datasets import make_classification 
```

```python
X, y = make_classification(n_samples=1000, n_features=4, n_informative=2, n_redundant=0, random_state=1, shuffle=False)
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Feature 1', fontsize = 15)
ax.set_ylabel('Feature 2', fontsize = 15)
ax.set_title('Generated Data', fontsize = 20)
ax.scatter(X[y==0,0], X[y==0,1], c='r', s=50)
ax.scatter(X[y==1,0], X[y==1,1], c='g', s=50)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
clf = AdaBoostClassifier(n_estimators=100) 
clf.fit(X_train, y_train)
print(clf.predict([[0, 4, 0, 0]]))
```

```python
ada_tr_score = clf.score(X_train, y_train)
ada_te_score = clf.score(X_test, y_test)
print(ada_tr_score,ada_te_score)
```

```python
from sklearn import tree
import graphviz 

decision_tree = tree.DecisionTreeClassifier()
decision_tree = decision_tree.fit(X_train, y_train)
tree_tr_score = decision_tree.score(X_train, y_train)
tree_te_score = decision_tree.score(X_test, y_test)
print(tree_tr_score,tree_te_score)
```

```python
graph_data = tree.export_graphviz(decision_tree, out_file=None) 
graph = graphviz.Source(graph_data) 
graph
```
